```python
#Adam을 사용할때는 스케줄링을 사용하지 않는다. 기본적으로 아담내에서는 러닝레이트를 조정하는 기능을 갖고 있다.
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
```

### 4) Scheduling

```python
optimizer = torch.optim.SGD(model.parameters(), lr=0.1, momentum=0.9)
#step_size = 30은 30epoch마다 러닝레이트를 조정한다는 의미이다.
#gamma=0.1은 옵티마이저에 선언된 러닝레이트에, 스케줄러에 선언된 스텝사이즈만큼의 에폭이 실행된 후 감마에 선언된 값이 러닝레이트와 곱해져서 다음 스탭의 러닝레이트로 사용된다.
scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=30, gamma=0.1)
```
